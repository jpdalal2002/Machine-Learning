# Problem Statement:

### Run the following Clustering Algorithms on bill_length and flipper_length attributes of 'penguins.csv' :
### 1) Agglomerative clustering (distance threshold = 100)
### 2) DBScan clustering (epsilon = 3, min point = 10)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
```

```python
dataset = pd.read_csv('penguins.csv')
```

```python
dataset
```

```python
print(dataset.shape)
```

```python
dataset = dataset[['bill_length_mm', 'flipper_length_mm']]
dataset = dataset.dropna(axis=0)
```

```python
dataset
```

### 1) Agglomerative clustering

```python
clusters = hierarchy.linkage(dataset, method="ward")

plt.figure(figsize=(16, 12))
dendrogram = hierarchy.dendrogram(clusters)
```

```python
plt.figure(figsize=(16,12))
plt.scatter(dataset['bill_length_mm'],dataset['flipper_length_mm'])
plt.xlabel('bill_length_mm')
plt.ylabel('flipper_length_mm')
plt.show()
```

```python
X = dataset.iloc[:,[0,1]].values
```

```python
X
```

```python
model = AgglomerativeClustering(distance_threshold=100,n_clusters=None)
```

```python
model.fit(X)
```

```python
y_pred = model.fit_predict(X)
```

```python
y_pred
```

```python
plt.figure(figsize=(16,12))
plt.scatter(dataset['bill_length_mm'],dataset['flipper_length_mm'],c=y_pred,cmap='rainbow')
plt.xlabel('bill_length_mm')
plt.ylabel('flipper_length_mm')
plt.show()
```

### 2) DBScan

```python
from sklearn.cluster import DBSCAN
```

```python
model1 = DBSCAN(eps=3,min_samples=10)
```

```python
model1.fit(X)
```

```python
y_pred = model1.fit_predict(X)
```

```python
y_pred
```

```python
plt.figure(figsize=(16,12))
plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='rainbow')
plt.show()
```
